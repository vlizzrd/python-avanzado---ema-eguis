```python
from gtts import gTTS
import os
from IPython.display import Audio, display
```

```python

texto_avoz = "Hola python texto a gtts"
tts = gTTS(text=texto_avoz, lang='es')


archivo_salida = "nota_voz.mp3"
tts.save(archivo_salida)
display(Audio(archivo_salida))
```
