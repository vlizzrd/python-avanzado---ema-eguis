```python
import pandas as pd
from datetime import datetime
from docxtpl import DocxTemplate
```

```python
for index, row in df.iterrows():
    datos_estudiante = {
        "nombre_y_apellido": row["Nombre y Apellido"],
        "asistencia": row["Asistencia"],
        "fecha_de_certificado": row["Fecha de certificado"],
        "nota": row["Nota"],
        "correo_electronico": row["Correo Electronico"],
        "curso": row["Curso"],
    }
    
    doc.render({**datos_instructor, **datos_estudiante})
    doc.save(f"Certificado_Plantilla - {row['Nombre y Apellido']}.docx")
```
